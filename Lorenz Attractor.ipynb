```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def lorenz_system(current, t, sigma, beta, rho):
    x, y, z = current
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]

# Parameters
sigma = 10.0
beta = 8.0/3.0
rho = 28.0

# Initial condition
initial = [1.0, 1.0, 1.0]

# Time points to solve at
t = np.linspace(0, 40, 10000)

# Solve the differential equations
solution = odeint(lorenz_system, initial, t, args=(sigma, beta, rho))

# Extract x, y, z from solution
x, y, z = solution.T

plt.rcParams['font.family'] = 'Serif'


# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor')

# Save the plot to a high-resolution PDF
fig.savefig("lorenz_attractor.pdf", format='pdf', dpi=300)

plt.show()
```
